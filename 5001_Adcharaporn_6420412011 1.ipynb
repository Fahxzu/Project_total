```python
#สร้างตัวแปร เพื่อหาค่าเฉลี่ยเงินเดือนแต่ละอาชีพ แล้วนำมา plot เพื่อเปรียบเทียบกัน 
# ok ใช้ stat    2 
lst_stu = []
lst_go = []
lst_work = []

salary1 = []
salary2 = []
salary3 = []

age1 = []
age2 = []
age3 = []

for doc in collection.find() :
    #print(doc['Work_type'])
    if (doc['Work_type']) == 'Student' :
        lst_stu.append(int(doc['Salary']))
        age1.append(int(doc['Age']))
    if (doc['Work_type']) == 'Government official' :
        lst_go.append(int(doc['Salary']))
        age2.append(int(doc['Age']))
    if (doc['Work_type']) == 'Office worker' :
        lst_work.append(int(doc['Salary']))
        age3.append(int(doc['Age']))
    
    #print(int(doc['Salary']))


mean_stu = statistics.mean(lst_stu)
mean_go = statistics.mean(lst_go)
mean_work = statistics.mean(lst_work)


print(mean_work)
```

```python
# creating the dataset   3 
y = [mean_stu,mean_go,mean_work]
x = ["Student","Government official","office worker"]
fig = plt.figure(figsize = (6, 5))
 
# creating the bar plot
plot = plt.bar(x, y, color ='orange',width = 0.3)
 
# Add the data value on head of the bar
for value in plot:
    height = value.get_height()
    plt.text(value.get_x() + value.get_width()/2.,
             1.002*height,'%d' % int(height), ha='center', va='bottom')    
    
plt.xlabel("Work_type",fontsize = 13)
plt.ylabel("Salary",fontsize = 13)
plt.title("Mean of salary by Occupation/Work_type")

# Display the graph on the screen
plt.show()


#เนื่องจากจำนวนข้อมูลที่สร้างขึ้นมีจำนวนค่อนข้างน้อย อาจจะสรุปได้ไม่แน่ชัด
#แต่ถ้าหากมองจากกราฟ สามารถกล่าวได้ว่า ค่า้ฉลี่ยนเงินเดือน
#ของอาชีพพนักงานของภาครัฐ จะมีรายได้สูงที่สุด รองลงมาคือพนักงเอกชน และสุดท้ายคือนักเรียน นักศึกษา
```
